Loading and merging the cvs files.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2015-01-01 to 2015-12-31.csv")
df2= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2016-01-01 to 2017-12-31.csv")
df3= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2018-01-01 to 2019-12-31.csv")
df4= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2020-01-01 to 2021-12-31.csv")
df5= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2022-01-01 to 2023-12-31.csv")

combined_df = pd.concat([df1,df2,df3,df4,df5],ignore_index= True)
```

```python
# Convert datetime column to datetime type
combined_df['datetime'] = pd.to_datetime(combined_df['datetime'])
print(combined_df.datetime)
```

Cleaning the data


```python
# Drop unnecessary columns
combined_df.drop(columns=["severerisk","name", "icon","stations","description"], inplace=True)
```

```python
# Fill missing values in preciptype
combined_df['preciptype'] = combined_df['preciptype'].fillna("none")
print(combined_df['preciptype'])
```

```python
#  Extract year and month

combined_df['year'] = combined_df['datetime'].dt.year
combined_df['month'] = combined_df['datetime'].dt.month
```

Chart 1: Average monthly temperature across the year (bar chart)

```python
monthly_avg_temp= combined_df.groupby('month')['temp'].mean()
monthly_avg_temp.index = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]

monthly_avg_temp.plot(
    kind= "bar",
    figsize= (8,5),
    title= "Avg monthly temp in dehradun(2015-2023)"
)

plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.yticks(
    [0, 5, 10, 15, 20, 25, 30, 35],
    ['0°C', '5°C', '10°C', '15°C', '20°C', '25°C', '30°C', '35°C']
)
plt.show()
```

Findings of Chart 1:
- This chart shows the average temperature of each month across 2015–2023.
- June is the hottest month with an average temperature of 27°C–28°C, while January is the coldest.
- Temperature rises steadily from January to June, then drops again — showing a clear seasonal pattern.

 Chart 2: Which month gets the most rainfall (bar chart).

```python
monthly_rainfall = combined_df.groupby('month')['precip'].mean()
monthly_rainfall.index = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]

max_month = monthly_rainfall.idxmax()
max_rainfall = monthly_rainfall.max()

print(
    f"The month with highest rainfall is {max_month}\n"
    f"with {max_rainfall: } mm of rainfall"
)

monthly_rainfall.plot(
    kind= 'bar',
    figsize=(8,5),
    title= "Most rainfall in the months"
)

plt.xlabel('Month')
plt.ylabel('Total Rainfall (mm)')
plt.tight_layout
plt.show()
```

Findings of Chart 2:
- This chart shows the average rainfall of each month across 2015–2023.
- July and August receive the highest average rainfall, which aligns with the Indian monsoon season that typically hits Dehradun between June and September.

Chart-3: Temperature Raise over the years


```python
temp_raise = combined_df.groupby('year')['temp'].mean()

temp_raise.plot(
    kind= 'line',
    figsize=(8,5),
    title= "Average Yearly Temperature"    
)

plt.xlabel('Year')
plt.ylabel('Tempeture (°C)')
plt.show()
```

Findings of Chart 3:
- This chart shows the average yearly temperature from 2015 to 2023.
- 2015 was the hottest year and 2020 was the coldest, possibly due to reduced traffic and industrial activity during the global lockdown.
- There is no consistent upward trend — temperatures fluctuate from year to year.

Chart 4: How many days per year have humidity above 80%?

```python
humidity_above_80 = combined_df[combined_df['humidity']>80]
high_humidity_days = humidity_above_80.groupby('year').size()

high_humidity_days.plot(
    kind= 'bar',
    figsize= (8,5),
    title= ' Number of days where Humidity is above 80%'
)

plt.xlabel('Year')
plt.ylabel('Number of days')
plt.show()
```

Findings of Chart 4:
- This chart shows how many days per year had humidity above 80%.
- Humid days vary significantly year to year, with 2021 having the most (around 160 days) and 2015 and 2023 having the fewest (around 60 days).
- There is no clear increasing or decreasing trend overall.

Chart 5: Average temperature range by month


```python
combined_df['temp_range'] = combined_df['tempmax']-combined_df['tempmin']
temp_range = combined_df.groupby('month')['temp_range'].mean()
temp_range.index=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]

temp_range.plot(
    kind= 'bar',
    figsize=(8,5),
    title= 'Average temperature range by month'
)
plt.xlabel('month')
plt.ylabel('Average temperature range by month')
plt.show()
```

Findings of Chart 5:
- This chart shows the average daily temperature range (tempmax − tempmin) for each month.
- Summer months like April and May have the largest swings, while monsoon months like July and August have the smallest.
- This is because monsoon clouds reduce daytime heat and keep nights warmer, narrowing the gap between daily high and low temperatures.
