Loading and merging the cvs files.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2015-01-01 to 2015-12-31.csv")
df2= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2016-01-01 to 2017-12-31.csv")
df3= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2018-01-01 to 2019-12-31.csv")
df4= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2020-01-01 to 2021-12-31.csv")
df5= pd.read_csv(r"D:\vs code\python\pandas\weather\data\dehradun 2022-01-01 to 2023-12-31.csv")

combined_df = pd.concat([df1,df2,df3,df4,df5],ignore_index= True)
```

```python
# Convert datetime column to datetime type
combined_df['datetime'] = pd.to_datetime(combined_df['datetime'])
print(combined_df.datetime)
```

Cleaning the data


```python
# Drop unnecessary columns
combined_df.drop(columns=["severerisk","name", "icon","stations","description"], inplace=True)
```

```python
# Fill missing values in preciptype
combined_df['preciptype'] = combined_df['preciptype'].fillna("none")
print(combined_df['preciptype'])
```

```python
#  Extract year and month

combined_df['year'] = combined_df['datetime'].dt.year
combined_df['month'] = combined_df['datetime'].dt.month
```

Chart 1: Average monthly temperature across the year (bar chart)

```python
monthly_avg_temp= combined_df.groupby('month')['temp'].mean()

monthly_avg_temp.plot(
    kind= "bar",
    figsize= (8,5),
    title= "Avg monthly temp in dehradun(2015-2023)"
)

plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.show()
```
